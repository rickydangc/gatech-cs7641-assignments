```python
import sklearn
from sklearn.metrics import *
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from scipy import stats
from scipy.stats import randint
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler

# Use scikit-learn to grid search the batch size and epochs
import numpy
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier

# Cross Validation
def five_fold_CV(clf, params, dx, dy):
    cv_model = RandomizedSearchCV(clf, params, scoring='f1', n_jobs=-1, 
                                  cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=13), 
                                  verbose=1, n_iter=5, refit=True)
    
    cv_model.fit(dx, dy.values.ravel())
    
    return cv_model.best_estimator_
```

```python
def logistic_regression_pred(X_train, Y_train, X_test):
    lr = LogisticRegression()
    tuned_parameters = {'max_iter': [100, 200, 500, 750, 1000], 
                    'tol': [0.0001, 0.001, 0.01, 0.1],
                    'C': [0.01, 0.1, 1.0, 5.0, 10.0, 25.0, 50.0, 100.0],
                    'solver': ['newton-cg'], 
                    'class_weight': [None, 'balanced']}
    
    best_lr_model = five_fold_CV(lr, tuned_parameters, X_train, Y_train)
    print("______________________________________________")
    print(best_lr_model)
    lr_y_test = best_lr_model.predict(X_test)
    lr_y_prob = best_lr_model.predict_proba(X_test)[:,1]
    return lr_y_test, lr_y_prob

def svm_pred(X_train, Y_train, X_test):
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    svm = SVC()
    tuned_parameters = {"C": [0.001, 0.01, 0.1, 2, 8, 32, 64, 128, 512, 1024, 2048],
                    'gamma':['scale', 'auto'],
                    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
                    'probability':[True], 
                    'tol': [0.1, 0.01, 0.001, 0.0001]}
    best_svm_model = five_fold_CV(svm, tuned_parameters, X_train, Y_train)
    print("______________________________________________")
    print(best_svm_model)
    svm_y_test = best_svm_model.predict(X_test)
    svm_y_prob = best_svm_model.predict_proba(X_test)[:,1]
    return svm_y_test, svm_y_prob

def knn_pred(X_train, Y_train, X_test):
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    neigh = KNeighborsClassifier()
    tuned_parameters = dict(n_neighbors=range(1, 30), weights=['uniform', 'distance'])
    best_knn_model = five_fold_CV(neigh, tuned_parameters, X_train, Y_train)
    print("______________________________________________")
    print(best_knn_model)
    neigh_y_pred = best_knn_model.predict(X_test)
    neigh_y_prob = best_knn_model.predict_proba(X_test)[:,1]
    return neigh_y_pred, neigh_y_prob

def decisionTree_pred(X_train, Y_train, X_test):
    tree = DecisionTreeClassifier()
    tuned_parameters = {"max_depth": [5, 10, 20, 50, 100],
              "max_features": [5, 10, 30, 100],
              "min_samples_leaf": [5, 10, 30, 100],
              "criterion": ["gini", "entropy"]}

    best_tree_model = five_fold_CV(tree, tuned_parameters, X_train, Y_train)
    print("______________________________________________")
    print(best_tree_model)
    tree_y_test = best_tree_model.predict(X_test)
    tree_y_prob = best_tree_model.predict_proba(X_test)[:,1]
    # get importance
    importances = best_tree_model.feature_importances_
    return tree_y_test, tree_y_prob, importances

def randomforest_pred(X_train, Y_train, X_test):
    rf = RandomForestClassifier()
    
    tuned_parameters = {'n_estimators':[50, 100, 250,  500, 750, 1000, 1250, 1500, 2000], 
                    'criterion':['gini', 'entropy'], 
                    'max_features':['log2', 'auto', None], 
                    'min_samples_split':[2, 3, 4], 
                    'max_depth': [3, 6, 9, 12, 15, 18, 21, 24, 32, None], 
                    'min_samples_leaf':[1, 2], 
                    'max_leaf_nodes': [None, 5, 10],
                    'min_impurity_decrease':[0.1, 0.01, 0.001, 0.0001, 0.00001],
                    'bootstrap': [True, False],
                    'class_weight': [None, 'balanced', 'balanced_subsample']}
    best_rf_model = five_fold_CV(rf, tuned_parameters, X_train, Y_train)
    print("______________________________________________")
    print(best_rf_model)
    forest_y_pred = best_rf_model.predict(X_test)
    forest_y_prob = best_rf_model.predict_proba(X_test)[:,1]
    importances = best_rf_model.feature_importances_
    return forest_y_pred, forest_y_prob, importances

def xgboost_pred(X_train, Y_train, X_test):
    xgb = GradientBoostingClassifier()
    
    tuned_parameters = {'n_estimators': stats.randint(150, 1000),
              'learning_rate': stats.uniform(0.01, 0.59),
              'max_depth': [3, 4, 5, 6, 7, 8, 9],
              "loss":["deviance"],
              "min_samples_split": np.linspace(0.1, 0.5, 12),
              "min_samples_leaf": np.linspace(0.1, 0.5, 12),
              "max_features":["log2","sqrt"],
              "criterion": ["friedman_mse",  "mae"],
              "subsample":[0.5, 0.618, 0.8, 0.85, 0.9, 0.95, 1.0]
             }
    best_xgb_model = five_fold_CV(xgb, tuned_parameters, X_train, Y_train)
    print("______________________________________________")
    print(best_xgb_model)
    xgb_y_pred = best_xgb_model.predict(X_test)
    xgb_y_prob = best_xgb_model.predict_proba(X_test)[:,1]
    importances = best_xgb_model.feature_importances_
    return xgb_y_pred, xgb_y_prob, importances

def nn_pred(X_train, Y_train, X_test):
# Function to create model, required for KerasClassifier
    def create_model(optimizer='adam'):
        # create model
        model = Sequential()
        model.add(Dense(12, input_dim=13, activation='relu'))
        model.add(Dense(1, activation='sigmoid'))
        # Compile model
        model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
        return model
    # fix random seed for reproducibility
    numpy.random.seed(13)
    # create model
    model = KerasClassifier(build_fn=create_model, epochs=100, batch_size=10, verbose=0)
    # define the grid search parameters
    optimizer = ['SGD', 'RMSprop', 'Adagrad', 'Adadelta', 'Adam', 'Adamax', 'Nadam']
    tuned_parameters = dict(optimizer=optimizer)
    best_nn_model = five_fold_CV(model, tuned_parameters, X_train, Y_train)
    print("______________________________________________")
    print(best_nn_model)
    nn_y_pred = best_nn_model.predict(X_test)
    nn_y_prob = best_nn_model.predict_proba(X_test)[:,1]    # summarize results

    return nn_y_pred, nn_y_prob

def feature_importance(importances):
    # summarize feature importance
    features=[]
    importance=[]
    for i,v in enumerate(importances):
        if v>0:
            features.append(X_train.columns[i])
            importance.append(v)
    fea_importance=pd.DataFrame({'features':features,'importance':importance}).sort_values(by='importance',ascending=False).reset_index(drop=True)
    return fea_importance

def classification_metrics(Y_pred, Y_true, Y_prob):
    #TODO: Calculate the above mentioned metrics
    #NOTE: It is important to provide the output in the same order
    acc = accuracy_score(Y_pred, Y_true)
    precision = precision_score(Y_pred, Y_true)
    recall = recall_score(Y_pred, Y_true)
    f1score = f1_score(Y_pred, Y_true)
    roc_auc_score=sklearn.metrics.roc_auc_score(Y_true, Y_prob)
    return acc,precision,recall,f1score, roc_auc_score

def display_metrics(classifierName,Y_pred,Y_prob,Y_true):
    print("______________________________________________")
    print(("Classifier: "+classifierName))
    acc, precision, recall, f1score ,roc_auc_score= classification_metrics(Y_pred,Y_true,Y_prob)
    print(("Accuracy: "+str(acc)))
    print(("Precision: "+str(precision)))
    print(("Recall: "+str(recall)))
    print(("F1-score: "+str(f1score)))
    print('ROC AUC Score: '+str(roc_auc_score))
    print("______________________________________________")
    print("")
```

```python
X = data.loc[:, data.columns != 'target']
y = data.loc[:, data.columns == 'target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13, shuffle=True, stratify=y)
```

```python
y_pred,y_prob = logistic_regression_pred(X_train,y_train,X_test)
display_metrics("Logistic Regression",y_pred,y_prob,y_test)
```

```python
y_pred,y_prob = svm_pred(X_train,y_train,X_test)
display_metrics("SVM",y_pred,y_prob,y_test)
```

```python
y_pred,y_prob,importances = decisionTree_pred(X_train,y_train,X_test)
display_metrics("Decision Tree",y_pred,y_prob,y_test)
```

```python
y_pred,y_prob,importances = randomforest_pred(X_train,y_train,X_test)
display_metrics("Random Forest",y_pred,y_prob,y_test)
```

```python
y_pred,y_prob,importances = xgboost_pred(X_train,y_train,X_test)
display_metrics("GradientBoostingClassifier",y_pred,y_prob,y_test)
```

```python
y_pred,y_prob = knn_pred(X_train,y_train,X_test)
display_metrics("K Nearest Neighbor",y_pred,y_prob,y_test)
```

```python
y_pred,y_prob = nn_pred(X_train,y_train,X_test)
display_metrics("Neural Networks",y_pred,y_prob,y_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import learning_curve
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np

def plot_learning_curve(pipeline):
    #
    # Use learning curve to get training and test scores along with train sizes
    #
    train_sizes, train_scores, test_scores, fit_times, _ = learning_curve(estimator=pipeline, X=X_train, y=y_train.values.ravel(),
                                                           cv=10, train_sizes=np.linspace(0.1, 1.0, 10), scoring='f1',
                                                         n_jobs=1, return_times=True)
    #
    # Calculate training and test mean and std
    #
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    fit_times_mean = np.mean(fit_times, axis=1)
    fit_times_std = np.std(fit_times, axis=1)
    #
    # Plot the learning curve 1
    #

    plt.plot(train_sizes, fit_times_mean, color='blue', marker='o', markersize=5, label='Training Accuracy')
    plt.fill_between(train_sizes, fit_times_mean + fit_times_std, fit_times_mean - fit_times_std, alpha=0.15, color='blue')
    plt.title("Scalability of the model")
    plt.xlabel('Training Data Size')
    plt.ylabel('Time')
    plt.grid()
    plt.legend(loc='lower right')
    plt.show()

    #
    # Plot the learning curve 2
    #
    plt.plot(train_sizes, train_mean, color='blue', marker='o', markersize=5, label='Training Accuracy')
    plt.fill_between(train_sizes, train_mean + train_std, train_mean - train_std, alpha=0.15, color='blue')
    plt.plot(train_sizes, test_mean, color='green', marker='+', markersize=5, linestyle='--', label='Validation Accuracy')
    plt.fill_between(train_sizes, test_mean + test_std, test_mean - test_std, alpha=0.15, color='green')
    plt.title('Learning Curve')
    plt.xlabel('Training Data Size')
    plt.ylabel('F1 Score')
    plt.grid()
    plt.legend(loc='lower right')
    plt.show()
```

#### Decision Tree

```python
# Need to use best params to plot leanring curve
# Create a pipeline; This will be passed as an estimator to learning curve method
pipeline = make_pipeline(StandardScaler(),
                        DecisionTreeClassifier(criterion='entropy', max_depth=20, max_features=5, min_samples_leaf=10))
plot_learning_curve(pipeline)
```

#### GBDT

```python
# Need to use best params to plot leanring curve
# Create a pipeline; This will be passed as an estimator to learning curve method
pipeline = make_pipeline(StandardScaler(),
                        GradientBoostingClassifier(learning_rate=0.031771046922684384, max_depth=8,
                           max_features='log2',
                           min_samples_leaf=0.13636363636363638,
                           min_samples_split=0.24545454545454548,
                           n_estimators=800, subsample=0.85))
plot_learning_curve(pipeline)
```

#### SVM

```python
# Need to use best params to plot leanring curve
# Create a pipeline; This will be passed as an estimator to learning curve method
pipeline = make_pipeline(StandardScaler(),
                        SVC(C=64, gamma='auto', kernel='linear', probability=True))
plot_learning_curve(pipeline)
```

#### KNN

```python
# Need to use best params to plot leanring curve
# Create a pipeline; This will be passed as an estimator to learning curve method
pipeline = make_pipeline(StandardScaler(),
                        KNeighborsClassifier(n_neighbors=14))
plot_learning_curve(pipeline)
```

#### NN

```python
def create_model(optimizer='adam'):
    # create model
    model = Sequential()
    model.add(Dense(12, input_dim=13, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# Need to use best params to plot leanring curve
# Create a pipeline; This will be passed as an estimator to learning curve method
pipeline = make_pipeline(StandardScaler(),
                        KerasClassifier(build_fn=create_model, epochs=100, batch_size=10, verbose=0))
plot_learning_curve(pipeline)
```
